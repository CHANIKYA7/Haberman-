### creating the box plot for age, year of treatment and positive lymph nodes

```python
plt.figure(figsize=[14,7])
plt.subplot(2,2,1)
sns.boxplot(data=h_man,x="age")
plt.subplot(2,2,2)
sns.boxplot(data=h_man,x= "year_of_treatment")
plt.subplot(2,2,(3,4))
sns.boxplot(data=h_man,x= "positive_lymph_nodes")
plt.show()
```
